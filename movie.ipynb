```
file_location = '/FileStore/tables/movies_metadata.csv'

file_type = 'csv'

infer_schema = 'false'

first_row_is_header = 'true'

delimiter = ','

movies = spark.read.format(file_type)\
.option('inferSchema', infer_schema)\
.option('header', first_row_is_header)\
.option('sep', delimiter)\
.load(file_location)

display(movies)
```

```
movies.write.format('parquet').option('header','true').save('dbfs:/FileStore/raw2/movies')
```

```
movies.printSchema()
```
