```python
## Relação de filmes com mais likes no facebook e seu respectivo lucro

%sql with profit_tbl as(
  select
    id_hashed as id_1,
    gross - budget as profit
  from
    df_financeiro
),
likes_tbl as (
  select
    id_hashed as id_2,
    movie_facebook_likes
  from
    df_facebook
),
movie_tbl as (
  select
    id_hashed as id_3,
    movie_title
  from
    df_info_movie
  where
    movie_title is not null
)
select
  movie_facebook_likes as likes_film,
  profit as profit,
  movie_title as movie_name
from
  likes_tbl
  inner join movie_tbl on likes_tbl.id_2 = movie_tbl.id_3
  inner join profit_tbl on likes_tbl.id_2 = profit_tbl.id_1
where
  profit is not null
order by
  likes_film desc
```

```python
## Receita por país

%sql with profit_tbl as (
  select
    id_hashed as id_1,
    gross - budget as profit
  from
    df_financeiro
),
countries_tbl as (
  select
    id_hashed as id_2,
    country
  from
    df_info_movie
)
select
  country as country,
  sum(profit) as profit
from
  profit_tbl
  inner join countries_tbl on profit_tbl.id_1 = countries_tbl.id_2
where
  profit is not null
group by
  country
order by
  profit desc
```

```python
# Diretores de filmes com piores receitas
%sql with profit_tbl as (
  select
    id_hashed as id_1,
    gross - budget as profit
  from
    df_financeiro
),
directors_tbl as (
  select
    id_hashed as id_2,
    director_name
  from
    df_cast
)
select
  director_name as director,
  sum(profit) as profit
from
  profit_tbl
  inner join directors_tbl on profit_tbl.id_1 = directors_tbl.id_2
where
  profit is not null

group by
  director
order by
  profit asc
```
